```python
import os
import pandas as pd

# This cell adds all the context of the word "help" from all Hansard files
# Directory containing all Hansard files
hansard_dir = "/Users/muhammadmushoffa/Desktop/corpus_linguistic/Hansard 2000 file sample"

def extract_from_all_files(directory, word_list):
    """Iterate through all .txt files in directory and extract word context"""
    output_file = f"/Users/muhammadmushoffa/Desktop/corpus_linguistic/help_all_files_context.csv"
    
    # Convert word list to lowercase once (moved outside loop)
    word_list_lower = [word.lower() for word in word_list]
    
    with open(output_file, 'w', encoding='utf-8') as outf:
        df = pd.DataFrame(columns=['left_context', 'target_word', 'right_context'])
        total_matches = 0
        
        for filename in sorted(os.listdir(directory)):
            if filename.endswith('.txt'):
                file_path = os.path.join(directory, filename)
                
                # Read the file 
                with open(file_path, 'r', encoding='utf-8') as f:
                    contents = f.read()
                
             
                
                contents = contents.replace('<speaker>', '')
                contents = contents.replace('</speaker>', '')
                contents = contents.split()

                
                # We find the word, then get 30 words before and 60 words after
                for i, content in enumerate(contents):
                    if content.lower() in word_list_lower:

                        start = max(0, i - 30)
                        left_context = ' '.join(contents[start:i])
                        end = min(len(contents), i + 61)
                        right_context = ' '.join(contents[i+1:end])
                        
                        # context = ' '.join(contents[i-30:end])
                        # outf.write(f"...{left_context} - {content} - {right_context}...\n")
                        new_row = pd.DataFrame([{'left_context': left_context, 'target_word': content, 'right_context': right_context}])
                        df = pd.concat([df, new_row], ignore_index=True)
                        
                        total_matches += 1
                
                df.to_csv(output_file, index=False, encoding='utf-8')
                


        print(f"\nTotal matches: {total_matches}")


# Extract "help" from all files
help_forms = [
    'help', 'helps', 'helped', 'helping', 'helpful', 'helpless', 
    'helpfully', 'helplessness', 'helper', 'helpeth', 'helpst', 'unhelpful'
]
extract_from_all_files(hansard_dir, help_forms)
```
