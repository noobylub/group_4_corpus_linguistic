```python
import re

# Read the two specific files
file1_path = "/Users/muhammadmushoffa/Desktop/corpus_linguistic/Hansard 2000 file sample/136872_1885-05-18_Supply—committee.txt"
file2_path = "/Users/muhammadmushoffa/Desktop/corpus_linguistic/Hansard 2000 file sample/136859_1885-05-18_Representation of the People and Registration of Voters (Ireland) Acts—duties of.txt"

output_file = "/Users/muhammadmushoffa/Desktop/corpus_linguistic/two_files_empty_line_split.txt"

def process_by_empty_lines(file_path, filename):
    with open(file_path, 'r', encoding='utf-8') as infile:
        content = infile.read()
    
    # print(f"\n=== Processing {filename} ===")
    # print(f"Original content:\n{repr(content)}\n")
    
    # Split by empty lines (one or more consecutive empty lines)
    # This matches \n\n or \n\s*\n patterns
    sections = re.split(r'\n\n+', content)
    
   
    return sections  # ADD THIS LINE

# ADD THESE LINES TO ACTUALLY CALL THE FUNCTION
sections1 = process_by_empty_lines(file1_path, "Supply—committee.txt")
print("-----")
print(sections1[1])
sections2 = process_by_empty_lines(file2_path, "Representation of the People...")
print("-----")
print(sections2[1])
```
