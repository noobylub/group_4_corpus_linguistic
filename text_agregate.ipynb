```python
import os
import re
import pandas as pd
import xml.etree.ElementTree as ET

# This cell adds all the context of the word "help" from all Hansard files
# Directory containing all Hansard files
hansard_dir = "/Users/muhammadmushoffa/Desktop/corpus_linguistic/Hansard 2000 file sample"

def extract_from_all_files(directory, word_list):
    """Iterate through all .txt files in directory and extract word context"""
    output_file = f"/Users/muhammadmushoffa/Desktop/corpus_linguistic/help_complete_lubbil.csv"
    
    # Convert word list to lowercase once (moved outside loop)
    word_list_lower = [word.lower() for word in word_list]
    
    df = pd.DataFrame(columns=['left_context', 'target_word', 'right_context', 'file_id', 'url', 'title', 'sitting_date', 'sitting_type', 'word_count'])
    total_matches = 0
    
    for filename in sorted(os.listdir(directory)):
        if filename.endswith('.txt'):
            file_path = os.path.join(directory, filename)
            
            # Read the file 
            with open(file_path, 'r', encoding='utf-8') as f:
                contents = f.read()
            
            parts = re.split(r"\n\s*\n", contents, maxsplit=1)
            
            contents = parts[1]
            metadata = parts[0]

            # Extracing KWIC 
            contents = contents.replace('<speaker>', '')
            contents = contents.replace('</speaker>', '')
            contents = contents.split()

            # Extract metadata first
            try:
                file_id_match = re.search(r'<FileID>(.*?)</FileID>', metadata)
                url_match = re.search(r'<URL>(.*?)</URL>', metadata)
                title_match = re.search(r'<Title>(.*?)</Title>', metadata)
                deb_ref_match = re.search(r'<DebRef>(.*?)</DebRef>', metadata)
                sitting_date_match = re.search(r'<SittingDate>(.*?)</SittingDate>', metadata)
                sitting_type_match = re.search(r'<SittingType>(.*?)</SittingType>', metadata)
                word_count_match = re.search(r'<WordCount>(.*?)</WordCount>', metadata)
                
                # extracts the capture text 
                file_id = file_id_match.group(1) if file_id_match else 'N/A'
                url = url_match.group(1) if url_match else 'N/A'
                title = title_match.group(1) if title_match else 'N/A'
                sitting_date = sitting_date_match.group(1) if sitting_date_match else 'N/A'
                sitting_type = sitting_type_match.group(1) if sitting_type_match else 'N/A'
                word_count = word_count_match.group(1) if word_count_match else 'N/A'
                
            except Exception as e:
                print(f"Error parsing metadata for {filename}: {e}")
                continue
            
            # We find the word, then get 30 words before and 60 words after
            for i, content in enumerate(contents):
                if content.lower() in word_list_lower:

                    start = max(0, i - 30)
                    left_context = ' '.join(contents[start:i])
                    end = min(len(contents), i + 61)
                    right_context = ' '.join(contents[i+1:end])
                    
                    new_row = pd.DataFrame([{
                        'left_context': left_context, 
                        'target_word': content, 
                        'right_context': right_context,
                        'file_id': file_id,
                        'url': url,
                        'title': title,
                        'sitting_date': sitting_date,
                        'sitting_type': sitting_type,
                        'word_count': word_count
                    }])
                    df = pd.concat([df, new_row], ignore_index=True)
                    
                    total_matches += 1

    # Write to CSV once at the end
    df.to_csv(output_file, index=False, encoding='utf-8')
    print(f"\nTotal matches: {total_matches}")
    print(f"CSV saved to: {output_file}")


# Extract "help" from all files
help_forms = [
    'help', 'helps', 'helped', 'helping', 'helpful', 'helpless', 
    'helpfully', 'helplessness', 'helper', 'helpeth', 'helpst', 'unhelpful'
]
extract_from_all_files(hansard_dir, help_forms)
```

```python
"""
Voice of help, passive (to be helped, be + VBN) vs. active (“normal” help). Consider 2
words to the left of help. Code as (Active, Passive). If help is not a verb, code as ‘NA’.
"""

import spacy
import pandas as pd
# IMPORTANT YOU NEED TO RUN THIS FIRST: python -m spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")  

df = pd.read_csv("help_complete_lubbil.csv")
# As shown in the documentation above, we check passivity by be + VBN
be_forms = {'be', 'am', 'is', 'are', 'was', 'were', 'been', 'being'}
df['Voice'] = 'NA'

for idx, row in df.iterrows():
    entire_context = row['left_context'] + ' ' + row['target_word'] + ' ' + row['right_context']
    doc = nlp(entire_context)
    for idx, token in enumerate(doc):
        if 'help' in token.text.lower():
            # Check if help is a verb
            if token.pos_ == 'VERB':
                # Check if it's passive (be + VBN)
                left_context = doc[max(0, idx-3):idx]
                if any(t.lemma_.lower() in be_forms for t in left_context) and token.tag_.startswith('VBN'):
                    df.at[idx, 'Voice'] = 'Passive'
                elif any(child.dep_ == "auxpass" for child in token.children) and token.tag_ == "VBN":
                    df.at[idx, 'Voice'] = 'Passive'
                else:
                    df.at[idx, 'Voice'] = 'Active'
            break

df.to_csv("help_complete_lubbil.csv", index=False)
```
